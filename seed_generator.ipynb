```python
whole_list = [daisy_list, dandelion_list, rose_list, sunflower_list, tulip_list]
```

```python
name_list = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']
```

```python
X_train_list = []
X_val_list = []
X_test_list = []

for flowers, name in zip(whole_list, name_list):
    X_train, X_test = train_test_split(flowers, test_size=0.15)
    X_train, X_val = train_test_split(X_train, test_size=0.175)
    X_train = [os.path.join(f'/C:/Users/stell/tutoring/Sarah/modularization/flowers/{name}', i) for i in X_train]
    X_val = [os.path.join(f'/C:/Users/stell/tutoring/Sarah/modularization/flowers/{name}', i) for i in X_val]
    X_test = [os.path.join(f'/C:/Users/stell/tutoring/Sarah/modularization/flowers/{name}', i) for i in X_test]
    X_train_list.extend(X_train) # list끼리 합쳐서 하나의 큰 리스트를 만드는 함수
    X_val_list.extend(X_val)
    X_test_list.extend(X_test)
```

```python
flower_list = ['train']+X_train_list+['valid']+X_val_list+['test']+X_test_list
```

```python
with open("./splits.txt", mode="w") as file:
    file.write('\n'.join(flower_list))
```
